```python
query = """
SELECT 
    name AS index_name,
    object_name(object_id) AS table_name,
    type_desc
FROM sys.indexes
WHERE object_id = object_id('dbo.listing_features');
"""

with engine.connect() as conn:
    result = conn.execute(text(query))
    indexes = result.fetchall()

# Display indexes nicely
import pandas as pd
pd.DataFrame(indexes, columns=['index_name', 'table_name', 'type_desc'])
```
